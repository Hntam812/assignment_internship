```python
import pandas as pd
import numpy as np
```

```python
data_path="D:/DataScience/Internship/assignments/formula.xlsx"
df=pd.read_excel(data_path,sheet_name='given')
df
```

```python

df["ADD"] = df["Param1"] + df["Param2"]
df["MINUS"] = df["Param2"] - df["Param1"]
df['DIV'] = np.where(df['Param2']==0, 0, df["Param1"] / df["Param2"])
df['FUNC2'] = np.where(df['Label'] == True, df['Param1']/df['Param2'], df[['Param1','Param2']].min(axis=1))
df['FUNC1'] = np.where(df['Label'] == True & (df['Param1'] > df['Param2']), df['Param1'], df[['Param1','Param2']].min(axis=1))

df
```

```python
from pandas import ExcelWriter
from pandas import ExcelFile
import openpyxl

with pd.ExcelWriter('formula.xlsx', mode='a', engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name="new_results")
```
